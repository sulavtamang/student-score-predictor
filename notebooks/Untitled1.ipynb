```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

student_scores = pd.read_csv('../datasets/study_hours_scores.csv')
student_scores
```

```python
# Split data into features and target
x = student_scores[['Study Hours']]  # Features
y = student_scores['Scores']  # Target

# Scaling the data using StandardScaler
# scaler = StandardScaler()
# data[['Study Hours', 'Scores']] = scaler.fit_transform(data[['Study Hours', 'Scores']])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Train a linear regression model
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)

# Predict the scores
y_pred = linear_model.predict(x_test)

# Calculate MSE and RMSE
mse = mean_squared_error(x_test, y_pred)
print(f'MSE: {mse:.2f}')

rmse = np.sqrt(mse)
print(f"RMSE: {rmse}")
```

```python
# Visualize the relationship
plt.scatter(student_scores['Study Hours'], student_scores['Scores'], alpha=0.5, color='b', edgecolors='k')
plt.plot(x_test, y_pred, color='red', linewidth=2)

plt.title("Study Hours vs Scores with Linear Regression Fit")
plt.xlabel("Study Hours (Scaled)")
plt.ylabel("Scores (Scaled)")
plt.show()
```
